```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('ggplot')
```

```python
# common plot formatting for below
def format_plot(ax, title):
    ax.xaxis.set_major_formatter(plt.NullFormatter())
    ax.yaxis.set_major_formatter(plt.NullFormatter())
    ax.set_xlabel('feature 1', color='gray')
    ax.set_ylabel('feature 2', color='gray')
    ax.set_title(title, color='gray')
```

```python
X, y = make_blobs(n_samples=10, n_features=2, centers=2, random_state=0)

print(X.shape)
X
```

```python
# plot the data
fig, ax = plt.subplots(figsize=(10, 6))
point_style = dict(cmap='Paired', s=50)
ax.scatter(X[:, 0], X[:, 1], c=y, **point_style)

# format plot
format_plot(ax, 'Input Data')
# ax.axis([-1, 4, -2, 7])

plt.show()
```

```python
X, y = make_blobs(n_samples=50, n_features=2, centers=3, random_state=0)
# cluster_std , standard deviation默认值为1

# plot the data
fig, ax = plt.subplots(figsize=(10, 6))
point_style = dict(cmap='rainbow', s=50)
ax.scatter(X[:, 0], X[:, 1], c=y, **point_style)

# format plot
format_plot(ax, 'cluster_std=1')
# ax.axis([-1, 4, -2, 7])

plt.show()
```

```python
X, y = make_blobs(n_samples=50, n_features=2, centers=3, random_state=0, cluster_std=0.3)

# plot the data
fig, ax = plt.subplots(figsize=(10, 6))
point_style = dict(cmap='rainbow', s=50)
ax.scatter(X[:, 0], X[:, 1], c=y, **point_style)

# format plot
format_plot(ax, 'cluster_std=0.3')
# ax.axis([-1, 4, -2, 7])

plt.show()
```
